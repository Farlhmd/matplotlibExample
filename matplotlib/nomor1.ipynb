```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
hasilXa = []
hasilYa = []
hasilXb = []
hasilYb = []
yTick = [-100, 0, 100, 200]
xTick = [-10, -5, 0, 5, 10]

for x in range(-10, 11, 1):
    fungsiA = (2*x*x)+(3*x)+5 
    fungsiB = (-2*x*x)+(4*x)+150
    hasilXa.append(x)
    hasilYa.append(fungsiA)
    hasilXb.append(x)
    hasilYb.append(fungsiB)
    
else:
    plt.xticks(xTick)
    plt.yticks(yTick)
    plt.plot(hasilXa, hasilYa, label = '2x^2+3x+5')
    plt.plot(hasilXb, hasilYb, label = '-2x^2+4x+150')
    plt.plot(0, 0,color='#ffffff', label = '-10≤x≤10 ')
    plt.title("grafik 2 fungsi")
    plt.xlabel("sumbu x")
    plt.ylabel("sumbu y")
    plt.legend()
    plt.show()
    
```
